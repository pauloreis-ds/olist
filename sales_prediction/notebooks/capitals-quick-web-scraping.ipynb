```python
import pandas as pd 
import requests 
from bs4 import BeautifulSoup

import os
```

```python
__file__ = "capitals-quick-web-scraping"
THIS_DIR = os.path.dirname(os.path.abspath(__file__))
BASE_DIR = os.path.dirname(os.path.dirname(os.path.dirname(os.path.abspath(__file__))))
DATA_DIR = os.path.join(os.path.join(BASE_DIR, 'data'))
```

```python
url="https://pt.wikipedia.org/wiki/Lista_de_capitais_do_Brasil_por_%C3%A1rea"
table_class="wikitable sortable jquery-tablesorter"
response=requests.get(url)
response.status_code
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
capitals_table=soup.find('table',{'class':"wikitable"})
```

```python
df=pd.read_html(str(capitals_table))
df=pd.DataFrame(df[0])
df
```

```python
df.loc[14,'Sede de governo'] = 'Salvador'
```

```python
df['Sede de governo'].to_csv(os.path.join(os.path.join(DATA_DIR, "sales_forecast_data"), "brazil_capitals.csv"))
```
